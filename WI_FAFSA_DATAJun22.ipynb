```python
%matplotlib inline
import leafmap
import pandas
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
WI_dis_fafsa = pandas.read_excel("WI Data/WI_DistrictLevel_data.xlsx") 
#data source: Form Your Future; link: https://formyourfuture.org/
WI_dis_fafsa #unfiltered fafsa data
```

```python
import geopandas as gpd #geopandas to read shape files
```

```python
SclDis_shp=r'WI Data/SchoolDis_ShapeFile/WI_Unified_School_Districts_2021.shp'
#data source: State Cartographer's Office; link: https://www.sco.wisc.edu/data/school-districts/
shp_dis=gpd.read_file(SclDis_shp)
print(type(shp_dis))
shp_dis  #unfiltered shpe file for school district maps.
```

```python
shp_dis.plot(cmap ='coolwarm') #school districts maps (working shape file)
```

```python
shp_dis.sort_values(by='LEAID', inplace=True)
shp_dis['LEAID'].astype(int)
shp_dis #sorted shape file ascending order of LEAID
```

```python
WI_dis_fafsa.sort_values(by='LEAID', inplace=True)
WI_dis_fafsa['LEAID'].astype(int)
WI_dis_fafsa #sorted xlx file ascending order fafsa data
```

```python
match_list=[]
for item in shp_dis['LEAID']:
    for i in WI_dis_fafsa['LEAID']:
        if int(i)==int(item):
            match_list.append(item)
print(len(match_list)) #findinf how many districts are common in both files. 
```

```python
nomatch_list=[]
for item in shp_dis['LEAID']:
    if item not in match_list:
        nomatch_list.append(item)
print(nomatch_list) #finding districts that are in shape file but not in xlx file. 

nomatch_list_xlx=[]
for item in WI_dis_fafsa['LEAID']:
    if item not in match_list:
        nomatch_list_xlx.append(item)
print(nomatch_list_xlx) #finding districts that are in xlx file but not in shape file 
```

```python
indexList=[]
for i in nomatch_list:
    a=shp_dis[shp_dis['LEAID']==i].index
    indexList.append(a)
print(indexList) #finding the indexes of the districts that are extra in shape file  
indexList_xlx=[]
for i in nomatch_list_xlx:
    a=WI_dis_fafsa[WI_dis_fafsa['LEAID']==i].index
    indexList_xlx.append(a)
print(indexList_xlx) #finding the indexes of the districts that are extra in xlx file
```

```python
indexList=[38,114,116,133,188,190,191,219,24,243,246,300,296,309,331,352]
shpM_sorted=shp_dis
for i in indexList:
    shpM_sorted=shpM_sorted.drop(index=i)
shpM_sorted #getting rid of the extra districts in shape file
```

```python
shpM_sorted.plot(cmap ='coolwarm') #shape file still functioning after modification. 
```

```python
indexList_xlx=[195,85,100,284,220,128,160,163,55,31,328,355]
xlxM_sorted=WI_dis_fafsa
for i in indexList_xlx:
    xlxM_sorted=xlxM_sorted.drop(index=i)
xlxM_sorted #getting rid of the extra districts in xlx file
```

```python
xlxM_sorted.columns
```

```python
xlxM_sorted[['rangefromJun22','rangetoJun22']]=xlxM_sorted['Estimated FAFSA\nCompletion Percentage\nJun10  2022'].str.split('-', expand=True)
#xlxM_sorted.drop(columns=['rangefromJun22'], inplace=True)
xlxM_sorted
```

```python
fromlists=[]
for i in xlxM_sorted['rangefromJun22']:
    fromlists.append(i)
shpM_sorted['rangefromJun22']=fromlists

tolists=[]
for i in xlxM_sorted['rangetoJun22']:
    tolists.append(i)
shpM_sorted['rangetoJun22']=tolists
shpM_sorted
#adding the fafsa data to the shape file data. 
```

```python
print(fromlists) #finding out the outliers
```

```python
shpM_sorted=shpM_sorted[shpM_sorted['rangefromJun22']!='<20']
shpM_sorted=shpM_sorted[shpM_sorted['rangefromJun22']!='>=80']
shpM_sorted=shpM_sorted[shpM_sorted['rangefromJun22']!='Missing/Insufficient Data']
shpM_sorted #getting rid of the outliers 
```
